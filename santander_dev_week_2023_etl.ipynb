```python
import pandas as pd

# 1. EXTRACT: Lendo o arquivo criado no Excel
df = pd.read_csv('/content/IDs(Planilha1).csv')
user_ids = df['id'].tolist()
print(f"IDs extraídos do Excel: {user_ids}")
```

```python
import json

# 1. Primeiro, carrega todos os dados do arquivo JSON para uma variável
with open('/content/users.json', 'r', encoding='utf-8') as f:
    dados_totais = json.load(f)

# 2. Função para buscar dentro da lista que acabou de carregar
def get_user(id):
    # Procura na lista de dados o usuário que tem o ID correspondente
    for user in dados_totais:
        if user['id'] == id:
            return user
    return None

users = [user for id in user_ids if (user := get_user(id)) is not None]

print(json.dumps(users, indent=2, ensure_ascii=False))
```
